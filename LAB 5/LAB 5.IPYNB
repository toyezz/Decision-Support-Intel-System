```python
kc_house_data.drop(['id','date','waterfront','zipcode','view','zipcode','lat','long'],axis=1,inplace=True)

print(kc_house_data)
```

```python
kc_house_data.isnull().sum()
```

```python
X=kc_house_data.drop(['price'],axis=1)
Y=kc_house_data['price']
```

```python
!pip install numpy
import numpy as np
kc_house_data.select_dtypes(np.number)
```

```python
X=kc_house_data.drop(['price'],axis=1)
Y=kc_house_data['price']
```

```python
X_train=np.array(X[0:int(0.80*len(X))])
Y_train=np.array(Y[0:int(0.80*len(Y))])
```

```python
X_test=np.array(X[int(0.80*len(X)):])
Y_test=np.array(Y[int(0.80*len(Y)):])
len(X_train),len(X_test),len(Y_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR_fit=LR.fit(X_train,Y_train)
LR_pred=LR_fit.predict(X_test)
from sklearn.metrics import accuracy_score
print("Logisitic Regression is %f percent accurate"%(accuracy_score(LR_pred,Y_test)*100))
```

```python
from sklearn.naive_bayes import GaussianNB
NB=GaussianNB()
NB_fit=NB.fit(X_train,Y_train)
NB_pred=NB_fit.predict(X_test)
from sklearn.metrics import accuracy_score
print("Naive Bayes is %f percent accurate"%(accuracy_score(NB_pred,Y_test)*100))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
seed=0
cv= KFold(n_splits=5,random_state=2, shuffle=True)
```

```python
def return_score(model,X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    return score
scores = []
```

```python
model = LogisticRegression()
for train_index, test_index in cv.split(X,Y):
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
    score = return_score(model,X_train, X_test, y_train, y_test)
    scores.append(score)
print("Accuracy score in each iteration: {}".format(scores))
print ("K-Fold Score: {}".format(np.mean(scores)))
```

```python
model = GaussianNB()
for train_index, test_index in cv.split(X,Y):
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
    score = return_score(model,X_train, X_test, y_train, y_test)
    scores.append(score)
print("Accuracy score in each iteration: {}".format(scores))
print("K-Fold Score: {}".format(np.mean(scores)))
```
